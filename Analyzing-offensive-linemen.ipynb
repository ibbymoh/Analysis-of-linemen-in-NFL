```python
"""
From the dataset description, we know that a pass block is anyone fully blocking a defender from the QB,
or anyone in a clear pass block stance. So we would like to filter the dataset based on those values.
"""
```

```python
df = df_pff[df_pff['pff_role']=='Pass Block']
```

```python
for col in df.columns:
    print(col)
    print(df[col].value_counts())
```

```python

fig, ax = pyplot.subplots(figsize=(60,10))
sns.countplot(ax=ax,x=df['pff_positionLinedUp'])
pyplot.xlabel('Position', fontsize=50);
pyplot.ylabel('Count', fontsize=50);
pyplot.title('Position played across all plays', fontsize=40)
pyplot.tick_params(axis='both', which='major', labelsize=30)

pyplot.show()
```
