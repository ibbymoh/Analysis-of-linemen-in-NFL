```python
new_df = pd.merge(df_plays[['gameId','playId','prePenaltyPlayResult','possessionTeam']],
                  df[filtered_cols],'left',left_on=['gameId','playId'],right_on=['gameId','playId'])
print(new_df)
```

```python
net_yardage = new_df.groupby(['gameId','possessionTeam'])['prePenaltyPlayResult'].sum()
print(net_yardage)
```

```python
net_yardage_1 = net_yardage.reset_index()
print(net_yardage_1.columns)
net_yardage_1 = net_yardage_1.groupby(['possessionTeam'])['prePenaltyPlayResult'].sum().reset_index()
fig, ax = pyplot.subplots(figsize=(60,30))
sns.barplot(x=net_yardage_1['possessionTeam'],y=net_yardage_1['prePenaltyPlayResult'])

pyplot.xlabel('Team', fontsize=50);
pyplot.ylabel('Total yards', fontsize=50);
pyplot.title('Yards per team in the first 8 weeks', fontsize=40)
pyplot.tick_params(axis='both', which='major', labelsize=30)
pyplot.show()
```

```python

hits_allowed_sum  = new_df.groupby(['gameId','possessionTeam'])['pff_hitAllowed'].sum()
beaten_by_defender_sum = new_df.groupby(['gameId','possessionTeam'])['pff_beatenByDefender'].sum()
hurryAllowed_sum = new_df.groupby(['gameId','possessionTeam'])['pff_hurryAllowed'].sum()
sackAllowed_sum = new_df.groupby(['gameId','possessionTeam'])['pff_sackAllowed'].sum()
backFieldBlock_sum = new_df.groupby(['gameId','possessionTeam'])['pff_backFieldBlock'].sum()

offensive_resistance = ((net_yardage**0.5) + backFieldBlock_sum )/(1+sackAllowed_sum+hurryAllowed_sum+beaten_by_defender_sum+hits_allowed_sum)
```

```python
hit_allowed_sum_1 =  new_df.groupby(['gameId','possessionTeam'])['pff_hitAllowed'].sum().reset_index()
hit_allowed_sum_1 = hit_allowed_sum_1.groupby(['possessionTeam'])['pff_hitAllowed'].sum().reset_index()

fig, ax = pyplot.subplots(figsize=(60,30))
print(hit_allowed_sum_1)
sns.barplot(x=hit_allowed_sum_1['possessionTeam'],y=hit_allowed_sum_1['pff_hitAllowed'])

pyplot.xlabel('Team', fontsize=50);
pyplot.ylabel('Total hits', fontsize=50);
pyplot.title('Hits', fontsize=40)
pyplot.tick_params(axis='both', which='major', labelsize=30)
pyplot.show()
```

```python
beaten_by_defender_sum_1 =  new_df.groupby(['gameId','possessionTeam'])['pff_beatenByDefender'].sum().reset_index()
beaten_by_defender_sum_1 = beaten_by_defender_sum_1.groupby(['possessionTeam'])['pff_beatenByDefender'].sum().reset_index()

fig, ax = pyplot.subplots(figsize=(60,30))
print(hit_allowed_sum_1)
sns.barplot(x=beaten_by_defender_sum_1['possessionTeam'],y=beaten_by_defender_sum_1['pff_beatenByDefender'])

pyplot.xlabel('Team', fontsize=50);
pyplot.ylabel('Total times', fontsize=50);
pyplot.title('Beaten by a defender', fontsize=40)
pyplot.tick_params(axis='both', which='major', labelsize=30)
pyplot.show()
```

```python
hurryAllowed_sum_1 =  new_df.groupby(['gameId','possessionTeam'])['pff_hurryAllowed'].sum().reset_index()
hurryAllowed_sum_1 = hurryAllowed_sum_1.groupby(['possessionTeam'])['pff_hurryAllowed'].sum().reset_index()

fig, ax = pyplot.subplots(figsize=(60,30))
sns.barplot(x=hurryAllowed_sum_1['possessionTeam'],y=hurryAllowed_sum_1['pff_hurryAllowed'])

pyplot.xlabel('Team', fontsize=50);
pyplot.ylabel('Total hurrys', fontsize=50);
pyplot.title('Hurrys', fontsize=40)
pyplot.tick_params(axis='both', which='major', labelsize=30)
pyplot.show()
```

```python
sackAllowed_sum_1 =  new_df.groupby(['gameId','possessionTeam'])['pff_sackAllowed'].sum().reset_index()
sackAllowed_sum_1 = sackAllowed_sum_1.groupby(['possessionTeam'])['pff_sackAllowed'].sum().reset_index()

fig, ax = pyplot.subplots(figsize=(60,30))
sns.barplot(x=sackAllowed_sum_1['possessionTeam'],y=sackAllowed_sum_1['pff_sackAllowed'])

pyplot.xlabel('Team', fontsize=50);
pyplot.ylabel('Total sacks', fontsize=50);
pyplot.title('Sacks', fontsize=40)
pyplot.tick_params(axis='both', which='major', labelsize=30)
pyplot.show()
```

```python
backFieldBlock_sum_1 =  new_df.groupby(['gameId','possessionTeam'])['pff_backFieldBlock'].sum().reset_index()
backFieldBlock_sum_1 = backFieldBlock_sum_1.groupby(['possessionTeam'])['pff_backFieldBlock'].sum().reset_index()

fig, ax = pyplot.subplots(figsize=(60,30))
print(hit_allowed_sum_1)
sns.barplot(x=backFieldBlock_sum_1['possessionTeam'],y=backFieldBlock_sum_1['pff_backFieldBlock'])

pyplot.xlabel('Team', fontsize=50);
pyplot.ylabel('Total backfield blocks', fontsize=50);
pyplot.title('Backfield blocks', fontsize=40)
pyplot.tick_params(axis='both', which='major', labelsize=30)
pyplot.show()
```

```python
sorted_off_res = off_res2.groupby(['possessionTeam']).mean().sort_values('off_res',axis=0,ascending=False)
sorted_off_res = sorted_off_res.reset_index()
# print(sorted_off_res.reset_index())
```

```python
# Offensive resistance by postions
print(new_df)

net_yardage_p = new_df.groupby(['pff_blockType'])['prePenaltyPlayResult'].sum()
print(net_yardage_p)

hits_allowed_sum_p  = new_df.groupby(['pff_blockType'])['pff_hitAllowed'].sum()
beaten_by_defender_sum_p = new_df.groupby(['pff_blockType'])['pff_beatenByDefender'].sum()
pff_hurryAllowed_sum_p = new_df.groupby(['pff_blockType'])['pff_hurryAllowed'].sum()
pff_sackAllowed_sum_p = new_df.groupby(['pff_blockType'])['pff_sackAllowed'].sum()
pff_backFieldBlock_sum_p = new_df.groupby(['pff_blockType'])['pff_backFieldBlock'].sum()

offensive_resistance_position = ((net_yardage_p)**0.5 +pff_backFieldBlock_sum_p )/(1+pff_sackAllowed_sum_p +pff_hurryAllowed_sum_p + beaten_by_defender_sum_p + hits_allowed_sum_p)
```

```python
off_res_p = offensive_resistance_position.to_frame().reset_index()
off_res_p.rename(columns ={0:'off_res'},inplace=True)
print(off_res_p)
```

```python
sns.barplot(x=off_res_p['pff_blockType'],y=off_res_p['off_res'])
pyplot.show()
```
